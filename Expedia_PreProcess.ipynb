```python
#Importing Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random
```

```python
#Defining the Path
file_path = r'D:\Kaggle\train.csv\train.csv'
```

```python
#Reading the Data
df = pd.read_csv(file_path)
```

```python
#Training data Set lengths(row)
len(df)
```

```python
#Dropping the Nulls from column orig_destination_distance
df_distance_drop_na = df.dropna(subset=['orig_destination_distance'])
```

```python
#df_distance_drop_na data Set lengths(row)
len(df_distance_drop_na)
```

```python
#Filtering the Data for users that have made the book
df_filtered_is_book = df_distance_drop_na[df_distance_drop_na['is_booking'] == 1]
```

```python
#df_distance_drop_na data Set lengths(row)
len(df_filtered_is_book)
```

```python
#Creating an attribute using check in and check out information
df_filtered_is_book['srch_ci'] = pd.to_datetime(df_filtered_is_book['srch_ci'])
df_filtered_is_book['srch_co'] = pd.to_datetime(df_filtered_is_book['srch_co'])

df_filtered_is_book['difference_in_days'] = (df_filtered_is_book['srch_co'] - df_filtered_is_book['srch_ci']).dt.days
```

```python
#Extracting Month information from user event date_time 
df_filtered_is_book['date_time'] = pd.to_datetime(df_filtered_is_book['date_time'])
df_filtered_is_book['month'] = df_filtered_is_book['date_time'].dt.month
```

```python
#Dropping the unecessary columns
columns_to_drop = ['date_time', 'srch_ci', 'srch_co']
df_filtered_is_book = df_filtered_is_book.drop(columns=columns_to_drop)
```
