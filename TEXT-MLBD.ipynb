<h2>PMI analysis</h2>

```python
import nltk
nltk.download('punkt')
```

```python
import requests
from lxml import html
import pandas as pd
pd.set_option('display.max_colwidth', 120)

import nltk
from nltk.collocations import *
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('averaged_perceptron_tagger')
stop_words = set(stopwords.words('english'))
stop_words.update([']', '['])
from IPython.display import display_html

def validType(ngram):#function to filter for ADJ/NN bigrams
    for word in ngram:
        if word in stop_words or word.isspace():
            return False
        
    tags = nltk.pos_tag(ngram)
    if tags[0][1] in ('JJ', 'JJR', 'JJS', 'NN', 'NNS', 'NNP', 'NNPS') and \
            tags[1][1] in ('NN', 'NNS', 'NNP', 'NNPS'): return True
    else: return False
    
def display_output(a, b):
    html_str = a.to_html(index=False) + b.to_html(index=False)
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
text = ' '.join(df_selected.head()['reviews.text'].astype(str))
```

```python
tokens = word_tokenize(text.lower())
bigrams = nltk.collocations.BigramAssocMeasures()
bigramsFinder = nltk.collocations.BigramCollocationFinder.from_words(tokens)
bigramsFinder.apply_freq_filter(3) #filter only those with more than 3 occurences
bigramsPMI = pd.DataFrame(list(bigramsFinder.score_ngrams(bigrams.pmi)), 
                          columns=['bigram','PMI']).sort_values(by='PMI', ascending=False)
filteredBigramsPMI = bigramsPMI[bigramsPMI.bigram.map(lambda x: validType(x))]
display_output(bigramsPMI, filteredBigramsPMI)
```

```python
#chi-sq test
bigramsChisq = pd.DataFrame(list(bigramsFinder.score_ngrams(bigrams.chi_sq)), 
                            columns=['bigram','chi-sq']).sort_values(by='chi-sq', ascending=False)
filteredBigramsChisq = bigramsChisq[bigramsChisq.bigram.map(lambda x: validType(x))]
display_output(bigramsChisq, filteredBigramsChisq)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import numpy as np

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
sw = set(stopwords.words('english'))
ps = PorterStemmer()
ls = WordNetLemmatizer()

def clean_text(review):
    ts = review.split()
    nts = [w.lower() for w in ts] #lower case words
    nts = [w for w in nts if w not in sw] #stop word removal
    nts = [ls.lemmatize(w) for w in nts] #apply lemmatization
    #nts = [ps.stem(w) for w in nts] #apply porter stemmer
    res = ' '.join(nts)
    return res
```

```python
corpus = ['Orion Corp reported a fall in its quarterly earnings :(', 
           'The store will open this fall in Bengaluru -:)', 
           'Adjustment to the fall in price level has been less effective !!!']
```

```python
vectorizer = TfidfVectorizer(lowercase=True, preprocessor=clean_text, ngram_range=(1,2), max_features=6)
X = vectorizer.fit_transform(corpus)
pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
from nltk.tokenize import word_tokenize, wordpunct_tokenize
nltk.download('punkt')
def custom_tokenizer(text):
    tokens = wordpunct_tokenize(text) #extract special characters
    return tokens

vectorizer = CountVectorizer(binary=True, preprocessor=clean_text, tokenizer=custom_tokenizer, max_features=15, token_pattern=None) #one hot encoding
X = vectorizer.fit_transform(corpus)
pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```
